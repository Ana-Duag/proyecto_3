```python
def analisis_eda(df):

    print("AN√ÅLISIS EXPLORATORIO DE DATOS\n")

    # Primeras filas
    print("Primeras 10 filas del DataFrame:")
    display(df.head(10))

    # Dimensiones
    print("\nDimensiones (filas, columnas):")
    print(df.shape)

    # Informaci√≥n general
    print("\nInformaci√≥n general:")
    df.info()

    # Tipos de datos
    print("\nTipos de datos por columna:")
    print(df.dtypes)

    # Estad√≠sticas num√©ricas
    print("\nDescripci√≥n de columnas num√©ricas:")
    display(df.describe())

    # Estad√≠sticas categ√≥ricas
    print("\nDescripci√≥n de columnas categ√≥ricas:")
    display(df.describe(include=['O']))

    # Valores nulos (conteo y porcentaje)
    print("\nValores nulos por columna (conteo y porcentaje):")
    nulls_count = df.isnull().sum()
    nulls_percent = (nulls_count / df.shape[0] * 100).round(2)  # nulos en porcentaje, redondeados a 2 decimales
    nulls_df = nulls_count.to_frame(name='Conteo')   #pasamos la serie a frame y le llamamos conteo
    nulls_df['Porcentaje'] = nulls_percent   # a√±adimos una columna con los porcentajes
    display(nulls_df[nulls_df['Conteo'] > 0].sort_values(by='Porcentaje', ascending=False))
    if nulls_count.sum() == 0:   
        print("No hay valores nulos en el DataFrame.")

    # Filas duplicadas
    print("\nFilas duplicadas:")
    dup_count = df.duplicated().sum()     #sumamos los duplicados
    print(f"Duplicadas: {dup_count}")
    if dup_count > 0:
        print("Duplicados (primeras 10 filas):")
        display(df[df.duplicated()].head(10))
    else:
        print("No hay filas duplicadas.")

    # Valores √∫nicos por columna
    print("\nValores √∫nicos por columna:")   #valores distintos en cada columna del DataFrame df
    unique_counts = df.nunique().sort_values(ascending=False)  # orden columnas de mayor a menor seg√∫n la cantidad de valores √∫nicos.
    display(unique_counts) 

    # Distribuci√≥n de variables categ√≥rica
    print("\nDistribuci√≥n de variables categ√≥ricas:")
    cat_cols = df.select_dtypes(include="object").columns
    for col in cat_cols:
        print(f"\nColumna: {col}")
        display(df[col].value_counts().head())

    # Estad√≠sticas b√°sicas de columnas num√©ricas
    print("\nEstad√≠sticas adicionales columnas num√©ricas:")
    num_cols = df.select_dtypes(include='number').columns
    # Comprobamos si hay columnas num√©ricas en el DataFrame
    # Sabemos que hay columnas num√©ricas, as√≠ que calculamos las estad√≠sticas directamente
    estadisticas_num = pd.DataFrame({
        'Media': df[num_cols].mean(),
        'Mediana': df[num_cols].median(),
        'Desviaci√≥n est√°ndar': df[num_cols].std(),
        'M√≠nimo': df[num_cols].min(),
        'M√°ximo': df[num_cols].max(),
        'Percentil 25': df[num_cols].quantile(0.25),   # a√±adido percentil 25
        'Percentil 75': df[num_cols].quantile(0.75)    # a√±adido percentil 75
    }).round(2)  # redondeamos todas las estad√≠sticas a 2 decimales para mayor legibilidad
    display(estadisticas_num)
```

```python
analisis_eda(df)
```
